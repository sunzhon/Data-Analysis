```python
import tensorflow as tf
import h5py

import json
import matplotlib.pyplot as plt

def plot_history(history_dict):
    print(history_dict.keys())
    plt.plot(history_dict['loss'],'r')
    plt.plot(history_dict['val_loss'],'g')
    plt.plot(history_dict['mae'])
    plt.legend(['train loss', 'valid loss','mae'])
    
    #plt.axis([0,150, 0.0,0.035])
    print('max train MAE: {:.4f} and max val MAE: {:.4f}'.format(max(history_dict['mae']),max(history_dict['val_mae'])))



base_folder='./models_parameters_results/2021-08-11/'
history_file=base_folder + 'training_051817/train_process/my_history_sub_1_2_3_4_5_6_7_8_9_10_11_12_13'

with open(history_file,'r') as fd:
    history_dict=json.load(fd)
    
plot_history(history_dict)

test_results_file=base_folder + '/test_051817/test_1/test_results.h5'
with h5py.File(test_results_file,'r') as fd:
    print(fd.keys())
    plt.figure(figsize=(16,10))
    plt.plot(fd['labels'][4200:4900,2],'r')
    plt.plot(fd['predictions'][4200:4900,2],'g')
    plt.legend(['true','prediction'])
```
